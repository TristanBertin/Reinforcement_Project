### La matrice de reward associe

```python
A = np.zeros((N,N))
A[-1,-1] = 10
A[0,-1] = 3
A
```


## State Value Function example

#### State Value Function : on evalue pour chaque tat la qualit de ce dernier en se basant sur la somme pondre des rewards futurs (les rewards sont eux donnees, mais peuvent etre sparses !) -> pas de politique, pas d'action value function, on calcule juste tous les state value function et on se dplace de sorte  maximiser  chaque instant cette fonction

```python
T = 10

def value(i,j,n):
    # Fonction qui parait complique mais qui est toute simple, 
    #tous les if sont juste l pour considrer les problmes aux bords du domaine
    # Pour chaque case, on value la somme pondre par les facteurs gamma des rewards...
    # T dsigne l'horizon, combien de pas de temps on regarde pour valuer un tat donn...
    
    if (i,j)==(N-1, N-1): #on est arriv au but, cela se finit ici
        return A[N-1, N-1]
    
    if n == T: # si on a atteint l'horizon max
        return A[i,j]
    
    else:
        n = n+1

        if i == 0:
            if j == 0 :
                return A[i,j] + gamma * (0.5 * value(i,j+1,n) + 0.5 * value(i+1, j,n))
            if j==N-1:
                return A[i,j] + gamma * (0.5 * value(i,j-1,n) + 0.5 * value(i+1, j,n))
            else:
                return A[i,j] + gamma * (0.333 * value(i,j-1,n) + 0.3333 * value(i, j+1,n) + 0.3333 * value(i+1, j,n))

        if i==N-1:
            if j == 0 :
                return A[i,j] + gamma * (0.5 * value(i-1,j,n) + 0.5 * value(i, j+1,n))
            elif j==N-1:
                return A[i,j] + gamma * (0.5 * value(i-1,j,n) + 0.5 * value(i, j-1,n))
            else:
                return A[i,j] + gamma * (0.333 * value(i,j-1,n) + 0.3333 * value(i, j+1,n) + 0.3333 * value(i-1, j,n))
        
        if j == 0:
            return A[i,j] + gamma * (0.333 * value(i,j+1,n) + 0.3333 * value(i+1, j,n) + 0.3333 * value(i-1, j,n))

        if j==N-1:
            return A[i,j] + gamma * (0.333 * value(i+1,j,n) + 0.3333 * value(i-1, j,n) + 0.3333 * value(i, j-1,n))

        else:
            return A[i,j] + gamma * (0.25 * value(i,j-1,n) + 0.25 * value(i, j+1,n) + 0.25 * value(i+1, j,n) +  0.25 * value(i-1, j,n))
```

#### Ce qui nous donne de manire rcursive la matrice des State Values !
(ie  combien un tat donne est susceptible de gnrer du reward futur)

```python
B = np.zeros((N,N))

for i in range(N):
    for j in range(N):
        B[i,j] = value(i,j,0)
        
np.around(B,1)  
```
